```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
home_dir = ".."

results_name = "paper_ablation_static"
df = pd.read_csv(f"{home_dir}/workflow/results/{results_name}/summary/results.csv")

results_name = "paper_ablation_static_5var"
df_5 = pd.read_csv(f"{home_dir}/workflow/results/{results_name}/summary/results.csv")

results_name = "paper_ablation_static_5k"
df_5k = pd.read_csv(f"{home_dir}/workflow/results/{results_name}/summary/results.csv")

results_name = "paper_ablation_static_10k"
df_10k = pd.read_csv(f"{home_dir}/workflow/results/{results_name}/summary/results.csv")

results_name = "paper_ablation_static_5k_500"
df_5k500 = pd.read_csv(f"{home_dir}/workflow/results/{results_name}/summary/results.csv")
```

```python
def plot_var(df, var, xlabel, mode='w'):
    df_local = df.copy()

    if mode == 'w':
        metric_prefix = 'w'
        ylabel_suffix = 'W'
        filename_prefix = 'window'
    else:
        metric_prefix = 's'
        ylabel_suffix = 'S'
        filename_prefix = 'summary'

    fig, axs = plt.subplots(1, 3, figsize=(12, 4), sharey=False)

    ax = axs[0]
    g = sns.lineplot(data=df_local, x=var, y=f'{metric_prefix}_f1', ax=ax, legend=False, err_style='bars', markers=True, ms=6, err_kws={'capsize':2.0, 'elinewidth': 0.5})
    g.set(xticks=df_local[var].unique())
    ax.set_xlabel(xlabel)
    ax.set_ylabel(f'F1$_{ylabel_suffix}$')

    ax = axs[1]
    g = sns.lineplot(data=df_local, x=var, y=f'{metric_prefix}_recall', ax=ax, legend=False, err_style='bars', markers=True, ms=6, err_kws={'capsize':2.0, 'elinewidth': 0.5})
    g.set(xticks=df_local[var].unique())
    ax.set_xlabel(xlabel)
    ax.set_ylabel(f'recall$_{ylabel_suffix}$')

    ax = axs[2]
    g = sns.lineplot(data=df_local, x=var, y=f'{metric_prefix}_precision', ax=ax, legend=False, err_style='bars', markers=True, ms=6, err_kws={'capsize':2.0, 'elinewidth': 0.5})
    g.set(xticks=df_local[var].unique())
    ax.set_xlabel(xlabel)
    ax.set_ylabel(f'precision$_{ylabel_suffix}$')

    plt.tight_layout()
    plt.savefig(f'{home_dir}/analysis/saved/ablation/{filename_prefix}_{var}.pdf', dpi=1000, bbox_inches='tight')
```

```python
# 2k samples
plot_var(df, 'n_ord', 'Number of orderings', 'w')
```

```python
# 5k samples
plot_var(df_5k, 'n_ord', 'Number of orderings', 'w')
```

```python
# 10k samples
plot_var(df_10k, 'n_ord', 'Number of orderings', 'w')
```

```python
# 5k samples, 500 repeats
plot_var(df_5k500, 'n_ord', 'Number of orderings', 'w')
```

```python
plot_var(df_5k500, 'n_ord', 'Number of orderings', 's')
```
