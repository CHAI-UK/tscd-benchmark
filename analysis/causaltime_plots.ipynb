```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
```

```python
results_name = "paper_causaltime"
home_dir = ".."
df = pd.read_csv(f"{home_dir}/workflow/results/{results_name}/summary/results.csv")
```

```python
datasets = ['medical', 'pm25', 'traffic']

df.loc[df['id'] == 'pcmci', 'id'] = 'PCMCI'
df.loc[df['id'] == 'tcdf', 'id'] = 'TCDF'

df.loc[df['id'] == 'cam-const', 'id'] = 'CAM-C'
df.loc[df['id'] == 'varlingam', 'id'] = 'VARLINGAM'
df.loc[df['id'] == 'dynotears', 'id'] = 'DYNOTEARS'
df.loc[df['id'] == 'diffan-const', 'id'] = 'DiffAN-C'
df.loc[df['id'] == 'timino', 'id'] = 'TiMINo'

algs = ['PCMCI', 'TCDF', 'CAM-C', 'VARLINGAM', 'DYNOTEARS', 'DiffAN-C', 'TiMINo', 'DOTS']
```

```python
def plot_dataset(df, algs, dataset):
    df_local = df.copy()
    df_local = df_local.loc[df_local['id'].isin(algs)]
    df_local = df_local.loc[df_local['seed'] == dataset]

    fig, axs = plt.subplots(1, 3, figsize=(12, 4), sharey=True)
    fig.suptitle(dataset)

    ax = axs[0]
    g = sns.barplot(data=df_local, x='s_f1', y='id', hue='id', orient='h', ax=ax, legend=False)
    g.set(xlabel='F1$_S$', ylabel='')

    ax = axs[1]
    g = sns.barplot(data=df_local, x='s_recall', y='id', hue='id', orient='h', ax=ax, legend=False)
    g.set(xlabel='recall$_S$', ylabel='')

    ax = axs[2]
    g = sns.barplot(data=df_local, x='s_precision', y='id', hue='id', orient='h', ax=ax, legend=False)
    g.set(xlabel='precision$_S$', ylabel='')

    plt.tight_layout()
    plt.savefig(f'{home_dir}/analysis/saved/causaltime/{dataset}.pdf', dpi=1000, bbox_inches='tight')

def plot_f1(df, algs, datasets, filename):
    df_local = df.copy()
    df_local = df_local.loc[df_local['id'].isin(algs)]

    fig, axs = plt.subplots(1, 3, figsize=(12, 4), sharey=True)

    for i, d in enumerate(datasets):
        df_dataset = df_local.loc[df_local['seed'] == d]
        ax = axs[i]
        g = sns.barplot(data=df_dataset, x='s_f1', y='id', hue='id', orient='h', ax=ax, legend=False)
        g.set(xlabel='F1$_S$', ylabel='', title=d)

    plt.tight_layout()
    plt.savefig(f'{home_dir}/analysis/saved/causaltime/{filename}.pdf', dpi=1000, bbox_inches='tight')
```

```python
df_local = df.copy()
df_local = df_local.loc[df_local['id'].isin(algs)]

fig, axs = plt.subplots(1, 3, figsize=(12, 4), sharey=True)

ax = axs[0]
g = sns.barplot(data=df_local, x='s_f1', y='id', hue='id', orient='h', ax=ax, legend=False)
g.set(xlabel='F1$_S$', ylabel='')

ax = axs[1]
g = sns.barplot(data=df_local, x='s_recall', y='id', hue='id', orient='h', ax=ax, legend=False)
g.set(xlabel='recall$_S$', ylabel='')

ax = axs[2]
g = sns.barplot(data=df_local, x='s_precision', y='id', hue='id', orient='h', ax=ax, legend=False)
g.set(xlabel='precision$_S$', ylabel='')

plt.tight_layout()
plt.savefig(f'{home_dir}/analysis/saved/causaltime/combined.pdf', dpi=1000, bbox_inches='tight')
```

```python
for d in datasets:
    plot_dataset(df, algs, d)
```

```python
plot_f1(df, algs, datasets, 'F1_S')
```
