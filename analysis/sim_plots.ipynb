```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
sns.color_palette("colorblind")
```

## Load data

```python
results_name = "paper_sim"
home_dir = ".."
df = pd.read_csv(f"{home_dir}/workflow/results/{results_name}/summary/results.csv")
```

## Renaming for nicer plots

```python
df.loc[df['id'] == 'diffan', 'id'] = 'DiffAN'
df.loc[df['id'] == 'diffan-const', 'id'] = 'DiffAN-C'
df.loc[df['id'] == 'DOTS-noconst', 'id'] = 'DOTS-noC' # without the temporal constraint
df.loc[df['id'] == 'DOTS', 'id'] = 'DOTS (ours)' # with the temporal constraint

df.loc[df['id'] == 'score', 'id'] = 'SCORE'
df.loc[df['id'] == 'das', 'id'] = 'DAS'
df.loc[df['id'] == 'cam', 'id'] = 'CAM'
df.loc[df['id'] == 'nogam', 'id'] = 'NoGAM'

df.loc[df['id'] == 'score-const', 'id'] = 'SCORE-C'
df.loc[df['id'] == 'das-const', 'id'] = 'DAS-C'
df.loc[df['id'] == 'cam-const', 'id'] = 'CAM-C'
df.loc[df['id'] == 'nogam-const', 'id'] = 'NoGAM-C'

df.loc[df['id'] == 'pcmci', 'id'] = 'PCMCI'
df.loc[df['id'] == 'pcmci+', 'id'] = 'PCMCI+'
df.loc[df['id'] == 'tcdf', 'id'] = 'TCDF'
df.loc[df['id'] == 'dynotears', 'id'] = 'DYNOTEARS'
df.loc[df['id'] == 'varlingam', 'id'] = 'VARLiNGAM'
df.loc[df['id'] == 'timino', 'id'] = 'TiMINo'
```

```python
df['id_gr'] = df['id']

df.loc[df['id_gr'] == 'DiffAN-C', 'id_gr'] = 'DiffAN'
df.loc[df['id_gr'] == 'DOTS-noC', 'id_gr'] = 'DOTS (ours)'
df.loc[df['id_gr'] == 'SCORE-C', 'id_gr'] = 'SCORE'
df.loc[df['id_gr'] == 'CAM-C', 'id_gr'] = 'CAM'
df.loc[df['id_gr'] == 'DAS-C', 'id_gr'] = 'DAS'
df.loc[df['id_gr'] == 'NoGAM-C', 'id_gr'] = 'NoGAM'

df['constraint'] = 'no'
df.loc[df['id'] == 'DiffAN-C', 'constraint'] = 'yes'
df.loc[df['id'] == 'DOTS (ours)', 'constraint'] = 'yes'
df.loc[df['id'] == 'CAM-C', 'constraint'] = 'yes'
df.loc[df['id'] == 'SCORE-C', 'constraint'] = 'yes'
df.loc[df['id'] == 'DAS-C', 'constraint'] = 'yes'
df.loc[df['id'] == 'NoGAM-C', 'constraint'] = 'yes'
```

```python
df['type'] = 'baseline'
df.loc[df['id'] == 'DOTS', 'type'] = 'ours'
```

```python
df['method'] = df['id']
```

```python
def plot_metric(df, algs, metric, ylabel):
    df_local = df.copy()
    df_local = df_local.loc[df_local['id'].isin(algs)]

    fig, axs = plt.subplots(1, 3, figsize=(12, 4), sharey=True)

    ax = axs[0]
    var = 'n'
    g = sns.lineplot(data=df_local, x=var, y=metric, hue='method', ax=ax, legend=False, err_style='bars', markers=True, style='method', ms=6, err_kws={'capsize':2.0, 'elinewidth': 0.5}, size='type', sizes={'baseline': 1.25, 'ours': 2.5}, palette='colorblind')
    g.set(xticks=df_local[var].unique())
    ax.set_xlabel('sample size')
    ax.set_ylabel(ylabel)

    ax = axs[1]
    var = 'features'
    g = sns.lineplot(data=df_local, x=var, y=metric, hue='method', ax=ax, legend=False, err_style='bars', markers=True, style='method', ms=6, err_kws={'capsize':2.0, 'elinewidth': 0.5}, size='type', sizes={'baseline': 1.25, 'ours': 2.5}, palette='colorblind')
    g.set(xticks=df_local[var].unique())
    ax.set_xlabel('number of features')

    ax = axs[2]
    var = 'n_lag'
    g = sns.lineplot(data=df_local, x=var, y=metric, hue='method', ax=ax, err_style='bars', markers=True, style='method', ms=6, err_kws={'capsize':2.0, 'elinewidth': 0.5}, size='type', sizes={'baseline': 1.25, 'ours': 2.5}, palette='colorblind')
    g.set(xticks=df_local[var].unique())
    ax.get_legend().remove()
    ax.set_xlabel('number of lags')

    fig.legend(loc='center left', bbox_to_anchor=(.99, .5), ncol=1, title=None, frameon=False)

    plt.tight_layout()
    plt.savefig(f'{home_dir}/analysis/saved/{metric}_all.pdf', dpi=1000, bbox_inches='tight')

def plot_metric_alt(df, algs, metric, ylabel, palette, style_order):
    df_local = df.copy()
    df_local = df_local.loc[df_local['id'].isin(algs)]

    fig, axs = plt.subplots(1, 3, figsize=(12, 4), sharey=True)

    ax = axs[0]
    var = 'n'
    g = sns.lineplot(data=df_local, x=var, y=metric, hue='method', ax=ax, legend=False, err_style='bars', markers=True, style='method', ms=6, err_kws={'capsize':2.0, 'elinewidth': 0.5}, palette=palette, hue_order=algs, style_order=style_order)
    g.set(xticks=df_local[var].unique())
    ax.set_xlabel('sample size')
    ax.set_ylabel(ylabel)

    ax = axs[1]
    var = 'features'
    g = sns.lineplot(data=df_local, x=var, y=metric, hue='method', ax=ax, legend=False, err_style='bars', markers=True, style='method', ms=6, err_kws={'capsize':2.0, 'elinewidth': 0.5}, palette=palette, hue_order=algs, style_order=style_order)
    g.set(xticks=df_local[var].unique())
    ax.set_xlabel('number of features')

    ax = axs[2]
    var = 'n_lag'
    g = sns.lineplot(data=df_local, x=var, y=metric, hue='method', ax=ax, err_style='bars', markers=True, style='method', ms=6, err_kws={'capsize':2.0, 'elinewidth': 0.5}, palette=palette, hue_order=algs, style_order=style_order)
    g.set(xticks=df_local[var].unique())
    ax.get_legend().remove()
    ax.set_xlabel('number of lags')

    fig.legend(loc='center left', bbox_to_anchor=(.99, .5), ncol=1, title=None, frameon=False)

    plt.tight_layout()
    plt.savefig(f'{home_dir}/analysis/saved/{metric}_all_alt.pdf', dpi=1000, bbox_inches='tight')

def plot_var(df, algs, var, xlabel, mode='w'):
    df_local = df.copy()
    df_local = df_local.loc[df_local['id'].isin(algs)]

    if mode == 'w':
        metric_prefix = 'w'
        ylabel_suffix = 'W'
        filename_prefix = 'window'
    else:
        metric_prefix = 's'
        ylabel_suffix = 'S'
        filename_prefix = 'summary'

    fig, axs = plt.subplots(1, 3, figsize=(12, 4), sharey=False)

    ax = axs[0]
    g = sns.lineplot(data=df_local, x=var, y=f'{metric_prefix}_f1', hue='method', ax=ax, legend=False, err_style='bars', markers=True, style='method', ms=6, err_kws={'capsize':2.0, 'elinewidth': 0.5}, size='type', sizes={'baseline': 1.25, 'ours': 2.5}, palette='colorblind')
    g.set(xticks=df_local[var].unique())
    ax.set_xlabel(xlabel)
    ax.set_ylabel(f'F1$_{ylabel_suffix}$')

    ax = axs[1]
    g = sns.lineplot(data=df_local, x=var, y=f'{metric_prefix}_recall', hue='method', ax=ax, legend=False, err_style='bars', markers=True, style='method', ms=6, err_kws={'capsize':2.0, 'elinewidth': 0.5}, size='type', sizes={'baseline': 1.25, 'ours': 2.5}, palette='colorblind')
    g.set(xticks=df_local[var].unique())
    ax.set_xlabel(xlabel)
    ax.set_ylabel(f'recall$_{ylabel_suffix}$')

    ax = axs[2]
    g = sns.lineplot(data=df_local, x=var, y=f'{metric_prefix}_precision', hue='method', ax=ax, err_style='bars', markers=True, style='method', ms=6, err_kws={'capsize':2.0, 'elinewidth': 0.5}, size='type', sizes={'baseline': 1.25, 'ours': 2.5}, palette='colorblind')
    g.set(xticks=df_local[var].unique())
    ax.get_legend().remove()
    ax.set_xlabel(xlabel)
    ax.set_ylabel(f'precision$_{ylabel_suffix}$')

    fig.legend(loc='center left', bbox_to_anchor=(.99, .5), ncol=1, title=None, frameon=False)

    plt.tight_layout()
    plt.savefig(f'{home_dir}/analysis/saved/{filename_prefix}_{var}.pdf', dpi=1000, bbox_inches='tight')
```
