```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../workflow/results/paper_ablation_lags/summary/results.csv")
```

```python
def plot_var(df, var, xlabel, mode='w'):
    df_local = df.copy()
    cap_s = 5.0

    if mode == 'w':
        metric_prefix = 'w'
        ylabel_suffix = 'W'
        filename_prefix = 'window'
    else:
        metric_prefix = 's'
        ylabel_suffix = 'S'
        filename_prefix = 'summary'

    fig, axs = plt.subplots(1, 3, figsize=(12, 4), sharey=True)

    ax = axs[0]
    df_select = df_local[df_local['seed'] == 'n_lag~1']
    g = sns.lineplot(data=df_select, x=var, y=f'{metric_prefix}_f1', ax=ax, legend=False, err_style='bars', markers=True, ms=6, err_kws={'capsize':cap_s, 'elinewidth': 0.5})
    g.set(xticks=df_local[var].unique())
    ax.set_xlabel('$\\tau_{max}$')
    ax.set_ylabel(f'F1$_{ylabel_suffix}$')
    ax.set_title('data lag = 1')
    ax.vlines(1, 0.3, 1.0, linestyles='dashed', colors='black')

    ax = axs[1]
    df_select = df_local[df_local['seed'] == 'n_lag~2']
    g = sns.lineplot(data=df_select, x=var, y=f'{metric_prefix}_f1', ax=ax, legend=False, err_style='bars', markers=True, ms=6, err_kws={'capsize':cap_s, 'elinewidth': 0.5})
    g.set(xticks=df_local[var].unique())
    ax.set_xlabel('$\\tau_{max}$')
    ax.set_ylabel(f'F1$_{ylabel_suffix}$')
    ax.set_title('data lag = 2')
    ax.vlines(2, 0.3, 1.0, linestyles='dashed', colors='black')

    ax = axs[2]
    df_select = df_local[df_local['seed'] == 'n_lag~3']
    g = sns.lineplot(data=df_select, x=var, y=f'{metric_prefix}_f1', ax=ax, legend=False, err_style='bars', markers=True, ms=6, err_kws={'capsize':cap_s, 'elinewidth': 0.5})
    g.set(xticks=df_local[var].unique())
    ax.set_xlabel('$\\tau_{max}$')
    ax.set_ylabel(f'F1$_{ylabel_suffix}$')
    ax.set_title('data lag = 3')
    ax.vlines(3, 0.3, 1.0, linestyles='dashed', colors='black', label='ground truth')

    plt.legend(loc='lower right')
    plt.tight_layout()
    plt.savefig(f'saved/ablation/n_lag.pdf', dpi=1000, bbox_inches='tight')
```

```python
plot_var(df, 'lag', 'max lag')
```
